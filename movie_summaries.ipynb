# Data exploration
This notebook contains some tests about data import and some data viz tools

```python
DATA_FOLDER = './data/'
MOVIES_FOLDER = DATA_FOLDER + 'movies_summaries/'
PLOT_SUMMARY_FOLDER = DATA_FOLDER + 'corenlp_plot_summaries'

CHARACTERS_FILE = MOVIES_FOLDER + 'character.metadata.tsv'
MOVIES_FILE = MOVIES_FOLDER + 'movies.metadata.tsv'
PLOT_SUMMARIES_FILE = MOVIES_FOLDER + 'plot_summaries.txt'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
summaries = pd.read_csv(PLOT_SUMMARIES_FILE, sep='\t', index_col=0, names=['plot'])
summaries.index.name = 'movie_id'
summaries
```

```python
with pd.option_context('display.max_colwidth', -1):

    car_synonyms = ['car', 'auto', 'jeep', 'limousine', 'pickup']

    display(summaries[summaries['plot'].str.contains('|'.join(car_synonyms), regex=True, case=False)].head(5))
```

```python
addiction_synonyms = ['drug', 'drugs', 'alcohol', 'beer', 'wine', 'cigarettes', 'cocain', 'whiksey', 'whisky', 'tequilla', 'vodka']

addiction_plots = summaries[summaries['plot'].str.contains('|'.join(addiction_synonyms), regex=True, case=False)]
```
